```python
#list Comprehension

#create a empty list
emptylist = [];

for list1 in range(1 , 1001):
    emptylist.append(list1)

print(emptylist)
print("---------------------")

empty_list = [];
for list2 in range(1 , 1001 , 2):
    empty_list.append(list2)
print(empty_list)

print("----------------------")

empty_list = [];

for list3 in range(1 , 1001):
    
    if(list3 % 2 == 0):
        
        empty_list.append(list3)
        
print(empty_list)

print("-----------------------")
```

```python


empty_list = [];

for list4 in range(1 , 1000):
    
    if(list4 % 2 != 0):
        
        empty_list.append(list4)
        
print(empty_list)

print("-------------------------------------")
```

```python
# Create a list using list comprehension that contains numbers from 1 to 1000 that are divisible by 8

empty_list = [];

for list5 in range(1 , 1000):
    
    if(list5 % 8 == 0):
        
        empty_list.append(list5)
        
print(empty_list)
```

```python
# Create a list using list comprehension that contains all those numbers that have a 5 in them.

empty_list = [];

for list6 in range(1 , 1001):
    
    if('5' in str(list6)):
        
         empty_list.append(list6)
            
print(empty_list)
```
